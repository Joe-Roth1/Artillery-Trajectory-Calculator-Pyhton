```python
# Remember the equation R= Vxt
# We know R should be about half a mile or about 804 meters
# We also know theta cannot be more than 45 degrees or pi/4 radians 
# Based on my data my average time is 1.8 seconds you can also round this to 2 just to see the difference 

import numpy as np
R= 804 # total distance in meters 
a= 1.7 # start time seconds
b=2 # end time seconds 
N=15 # number of iterations 
# formulas 
xi = np.arange(a,b) # arrange accordingly
t_Estimates =np.linspace(a,b,N+1) # array list of possible times in seconds 
g=-9.81 # gravity m/s^2  
Vx=R/t_Estimates # horizontal velocity
Vy =(1/2)*g*t_Estimates # vertical velocity 
V=np.sqrt(Vy**2+Vx**2) # initial velocity 
# to find theta
theta = np.arctan(Vy/Vx)
print('Estimations of theta:' , abs(np.degrees(theta)), 'degrees')
print('\n')
print('Estimations of initial velocity:', V, 'm/s')
```
